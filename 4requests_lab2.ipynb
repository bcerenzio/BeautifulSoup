```python
import numpy as np; import pandas as pd; from bs4 import BeautifulSoup as bs
import re; from bs4 import Comment; from bs4 import Tag; import requests
```

```python
r = requests.get('https://www.baseball-reference.com/players/c/colege01.shtml')
print(r.text)    
```

```python
with open('gerrit_request.txt', 'w', encoding = 'utf-8') as f:
    f.write(r.text)
```

```python
with open('gerrit_request.txt', encoding = 'utf-8') as f:
    soup = bs(f)

type(soup)
```

```python
comments = soup.find_all(string = lambda text: isinstance(text,Comment))
comments_str = ''.join(comments)
comments1 = bs(comments_str)
```

```python
comments1
```

```python
comments1.find_all(id = 'div_pitching_advanced')
```

```python
gerrit_advanced_pitching = comments1.find_all(id = 'div_pitching_advanced')
```

```python
len(gerrit_advanced_pitching)
```

```python
help(re.compile)
```

```python
cols = []
h = gerrit_advanced_pitching[0].find_all('thead')[0].find_all('th', {'scope':re.compile('col')})
for i in h:
    print(i.string)
    cols.append(i.string)
```

```python
body = gerrit_advanced_pitching[0].find_all('tbody')
cells = []
for i in body[0].find_all(re.compile('th|td')):
    print(i.get_text())
    cells.append(i.get_text())
```

```python
num_cols = len(cols)
num_cols
```

```python
cells1 = np.array(cells).reshape(int(len(cells)/num_cols),num_cols)
cells1
```

```python
gerrit_cole_advanced_pitching = pd.DataFrame(cells1, columns = cols)
gerrit_cole_advanced_pitching
```
